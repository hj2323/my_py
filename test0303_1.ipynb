```python
from bs4 import BeautifulSoup
import urllib.request as req
```

```python
def content(url):
    res1=req.urlopen(url)
    bs1=BeautifulSoup(res1, 'html.parser')
    content=bs1.select('#mw-content-text > div.mw-parser-output > div > p')
    print(str(content).replace('[]','').replace('<p>','').replace('<br />',''))
```

```python
url='https://ko.wikisource.org/wiki/%EC%A0%80%EC%9E%90:%EC%9C%A4%EB%8F%99%EC%A3%BC'
res=req.urlopen(url)
#print(res.read().decode('utf-8'))
bs=BeautifulSoup(res, 'html.parser')
li_list=bs.select('#mw-content-text > div.mw-parser-output > ul > li')

for li in li_list:
    print(li.a.string)
    sub_list=li.find_all('li')
    if len(sub_list)==0:
        url='https://ko.wikisource.org'+li.a['href']
        content(url)
    for sub in sub_list:
        print('-',sub.a.string)
        url='https://ko.wikisource.org'+sub.a['href']
        content(url)
```
