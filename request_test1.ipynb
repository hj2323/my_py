```python
#웹 콘텐츠 요청-Urllib패키지 활용
import urllib.request
from urllib.request import urlopen
```

```python
res=urllib.request.urlopen("http://www.naver.com")
print(type(res))
print(res.status)
print(res.version)
print(res.msg)
```

```python
res_header=res.getheaders()
print("[header 정보]")
for header in res_header:
    print(header)
```

```python
#예외 처리 방법
from urllib.error import HTTPError
from urllib.error import URLError
```

```python
try:
    html=urlopen("https://java001.com")
except HTTPError as e:
    print("HTTP 에러입니다")
except URLError as e:
    print("존재하지 않는 사이트입니다")
else:
    print(html.read())
```

```python
url="http://unico2013.dothome.co.kr/crawling/tagstyle.html"
res=urlopen(url)
encoding=res.info().get_content_charset()
print(res)
print('[header 정보]')
res_header=res.getheaders()


for header in res_header:
    print(header)
print(res.read().decode('utf-8'))
```

```python
import urllib.request
print("================================================")
url='http://www.python.org/'
f=urllib.request.urlopen(url)
print(type(f))
print(type(f.info()))
encoding=f.info().get_content_charset()
print(url, '페이지의 인코딩 정보:' , encoding)
text = f.read(500).decode(encoding)
print(text)
```

```python
#매개변수를 추가하여 인터넷 리소스를 요청하는 방법
import urllib.parse
```

```python
api="https://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp"
values={'stnid':'108'}
params=urllib.parse.urlencode(values)
url=api+'?'+params
print('url=', url)
data=urllib.request.urlopen(url).read()
text=data.decode('utf-8')
print(text)
```

```python
#request 패키지 활용
params=urllib.parse.urlencode({'category':'역사','page':25})
url="http://unico2013.dothome.co.kr/crawling/exercise.php?%s" % params 
with urllib.request.urlopen(url) as f:
    print(f.read().decode('utf-8'))
```
