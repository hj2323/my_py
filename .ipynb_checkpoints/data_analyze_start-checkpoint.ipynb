```python
import numpy as np
import pandas as pd
import matplotlib
```

```python
#numpy
#numpy 배열 만들기
data=np.random.rand(50)
print(type(data))
print(data)
```

```python
#pandas
seri=pd.Series(data)#Series : 인덱스를 가지고 있는 일차원 배열
print(type(seri))
print(seri)
```

```python
seri.plot()
```

```python
matplotlib.style.use('ggplot')
seri.plot()
```

```python
#이차원배열
data_set=np.random.rand(10,3)
print(data_set)
print(type(data_set))
print(data_set.shape)
```

```python
df=pd.DataFrame(data_set, columns=['A','B','C'])
print(df)
print(type(df))
df
```

```python
df.plot()#데이터 프레임으로 차트 그리기 - 기본형 라인
```

```python
df.plot(kind='bar')
```

```python
df.plot(kind='barh')
```

```python
df.plot(kind='bar', stacked=True)
```

```python
df.plot(kind='area', stacked=False)
#df.plot(kind='area', stacked=True)
```

```python
seri2=pd.Series(np.random.rand(5), index=['A','B','C','D','E'],name='Series')
seri2.plot(kind='pie',autopct='%.2f',fontsize=20,figsize(8,6))
```
