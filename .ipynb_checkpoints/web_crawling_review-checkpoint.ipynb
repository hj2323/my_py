```python
import urllib.parse
```

```python
api='http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp'
value={'stnid':'108'}
data=urllib.parse.urlencode(value)
url=api+'?'+data
text=urllib.request.urlopen(url).read().decode('utf-8')
print(text)
```

```python
import requests
url=api
res=requests.get(api,value)
print(type(res))
print(res.text)#-->html
```
